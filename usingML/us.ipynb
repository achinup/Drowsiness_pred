```python
import os
import shutil
import pandas as pd

# Load CSV file
csv_path = "_annotations.csv"  # Change this to your CSV file path
df = pd.read_csv(csv_path)

# Define source folder where images are stored
source_folder = "train/"  # Change this to the folder where your images are stored

# Define target folders
drowsy_folder = "drowsy_images/"
awake_folder = "awake_images/"

# Create folders if they don't exist
os.makedirs(drowsy_folder, exist_ok=True)
os.makedirs(awake_folder, exist_ok=True)

# Iterate over each row in the CSV
for index, row in df.iterrows():
    filename = row["filename"]  # Extract filename
    class_label = row["class"]  # Extract class (drowsy/awake)
    
    # Define source and destination paths
    source_path = os.path.join(source_folder, filename)
    
    if class_label == "drowsy":
        destination_path = os.path.join(drowsy_folder, filename)
    elif class_label == "awake":
        destination_path = os.path.join(awake_folder, filename)
    else:
        continue  # Skip if class is not recognized

    # Move the file to the corresponding folder
    if os.path.exists(source_path):  # Check if file exists before moving
        shutil.move(source_path, destination_path)
        print(f"Moved {filename} to {destination_path}")
    else:
        print(f"File {filename} not found in {source_folder}")

print("Image sorting complete! ")
```
